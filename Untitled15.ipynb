```python
from experta import *

# تعريف الـ Facts المستخدمة في النظام
class Environment(Fact): pass   # بيئة العمل
class Job(Fact): pass           # نوع المهمة
class Feedback(Fact): pass      # هل هناك تغذية راجعة
class Medium(Fact): pass        # الوسيلة التعليمية الناتجة
class Urgency(Fact): pass       # مستوى الاستعجال

# محرك النظام الخبير
class MediaAdvisorFinal(KnowledgeEngine):

    # تحديد نوع البيئة
    @Rule(Environment(environment='machines'))
    def physical_object(self):
        self.declare(Fact(stimulus_situation='physical object'))

    # مهمة الإصلاح تحتاج تعلم عملي
    @Rule(Job(job='repairing'))
    def hands_on(self):
        self.declare(Fact(stimulus_response='hands-on'))

    # القاعدة الأساسية (أولوية منخفضة)
    @Rule(Fact(stimulus_situation='physical object'),
          Fact(stimulus_response='hands-on'),
          Feedback(feedback='required'),
          salience=0)
    def workshop(self):
        print("Result: Workshop selected")
        self.declare(Medium(medium='workshop'))

    # قاعدة الطوارئ (أولوية عالية)
    @Rule(Fact(stimulus_situation='physical object'),
          Fact(stimulus_response='hands-on'),
          Urgency(level='high'),
          salience=100)
    def emergency_rule(self):
        print("Result: Digital Guide selected due to high urgency")
        self.declare(Medium(medium='digital_guide'))
        self.halt()
```

```python
engine = MediaAdvisorFinal()
engine.reset()

engine.declare(
    Environment(environment='machines'),
    Job(job='repairing'),
    Feedback(feedback='required'),
    Urgency(level='high')
)

engine.run()
```
